#### 12. Write a short program that prints the numbers 1 to 10 using a for loop. Then write an equivalent program that prints the numbers 1 to 10 using a while loop.

```python
#Answer:-
def forloop_gen10():
    """This function generates range from 0 to 10 by using for loop"""
    for i in range(1,11):
        print(i,end='')
        
def whileloop_gen10():
    """This function generates range from 0 to 10 by using while loop"""
    a=0
    while a < 10:
        a=a+1
        print(a,end='')
```

```python
forloop_gen10()
```

```python
whileloop_gen10()
```
