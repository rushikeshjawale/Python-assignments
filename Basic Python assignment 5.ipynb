## Q3 What is the most significant distinction between a dictionary and a list?

#### Answer:
• Dictionary stores data in key-value format hence to acess data of dictionary we have to call key of dictionary of that specific value. Where in list data is stored directly i.e without using keys, hence we call acess data from list by calling a specific index.

• The items stored in a dictionary are unordered, while the items in a list are ordered.

### 6. If a dictionary is stored in spam, what is the difference between the expressions 'cat' in spam and 'cat' in spam.values?
#### Answer:

'cat' in spam checks whether there is a 'cat' key in the dictionary, while 'cat' in spam.values() checks whether there is a value 'cat' for one of the keys in spam.
