#### 5. How many global scopes are there in a Python program? How many local scopes?
Global scope:- Whenever a variable is defined outside any function, it becomes a global variable, and its scope is anywhere within the program. Which means it can be used by any function.

Local scope:- Whenever you define a variable within a function, its scope lies ONLY within the function. It is accessible from the point at which it is defined until the end of the function and exists for as long as the function is executing.

#### 6. What happens to variables in a local scope when the function call returns?
